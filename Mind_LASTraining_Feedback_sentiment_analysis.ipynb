```python
from textblob import TextBlob

# Define a function for sentiment analysis
def analyze_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    sentiment = "Positive" if polarity > 0 else "Negative" if polarity < 0 else "Neutral"
    return polarity, sentiment

# List of columns to analyze (assuming these are relevant to sentiment analysis)
columns_to_analyze = ["Feeback_To_Trainer", "Feedback_To_Facilitators", "Suggestions"]

# Apply sentiment analysis to each column
for column in columns_to_analyze:
    comment[f"{column}_Polarity"], comment[f"{column}_Sentiment"] = zip(
        *comment[column].apply(analyze_sentiment)
    )

# Preview the updated DataFrame
print(comment.head())

# Visualize the sentiment distribution
for column in columns_to_analyze:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=comment, x=f"{column}_Sentiment", palette="viridis", order=["Positive", "Neutral", "Negative"])
    plt.title(f"Sentiment Distribution for {column}")
    plt.xlabel("Sentiment")
    plt.ylabel("Count")
    plt.show()

# Save the updated DataFrame to a new Excel file
output_path = 'Feedback_Analysis_Results.xlsx'
comment.to_excel(output_path, index=False)

print(f"Sentiment analysis completed and saved to {output_path}")
```

```python
from collections import Counter
from wordcloud import WordCloud

#Split and count occurrences of fields
fields_series = comment["What fields are you interested in for our future training?"].dropna().str.split(", ")
fields_flat = [field.strip() for sublist in fields_series for field in sublist]
field_counts = Counter(fields_flat)

fields_df = pd.DataFrame(field_counts.items(), columns=["Field", "Count"]).sort_values(by="Count", ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=fields_df, x="Count", y="Field", palette="viridis")
plt.title("Fields of Interest for Future Training")
plt.xlabel("Count")
plt.ylabel("Field")
plt.show()
```

```python
# Extract the day of the week using string operations
comment["Day of Week"] = comment["Favourite day"].str.extract(r'\((.*?)\)')
print(comment["Day of Week"].head())
```
